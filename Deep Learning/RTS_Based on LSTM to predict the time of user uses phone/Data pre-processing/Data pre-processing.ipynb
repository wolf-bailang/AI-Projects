```python
import csv
import json
#from pandas import read_csv
from datetime import datetime
#from matplotlib import pyplot as plt

#num, year, month, day, week, hour1, min1, sec1, p, hour2, min2, sec2=0
dataset=[]
year=0
month=0
day=0
week=0
hour1=0
min1=0
sec1=0
hour2=0
min2=0
sec2=0
dataset.append(['year', 'month', 'day', 'week', 'hour', 'min', 'sec', 'p'])

# load data
filename = '0.csv'
```

```python
def x(row):    
    num=int(row[1])
    hour1=int(row[2])
    min1=int(row[3])
    sec1=int(row[7])
    hour2=int(row[9])
    min2=int(row[10])
    sec2=int(row[11])
    t=0
    while(t==0):
        dataset.append([num, hour1, min1, sec1, 1])
        if(hour1==hour2 and min1==min2 and sec1==sec2):
            #t=1
            break
        else:          
            sec1+=1
            if(sec1 == 60):
                min1+=1
                sec1=0
            if(min1 == 60):
                hour1+=1
                min1=0
            if(hour1 == 24):
                hour1=0
```

```python
hour=0
mine=0
sec=0

with open(filename,encoding='UTF-8') as f:
    reader = csv.reader(f)
    header_row = next(reader)
    for row in reader:
        year=int(row[1])
        month=int(row[2])
        day=int(row[3])
        week=int(row[4])
        hour1=int(row[5])
        min1=int(row[6])
        sec1=int(row[7])
        hour2=int(row[9])
        min2=int(row[10])
        sec2=int(row[11])
        t=0
        while(t==0):
            dataset.append([num, hour, mine, sec, 0])
            if(hour==hour1 and mine==min1 and sec==sec1):
                x(row)
                hour=hour2
                mine=min2
                sec=sec2+1
                t=1
            else:   
                sec+=1
                if(sec == 60):
                    mine+=1
                    sec=0
                if(mine == 60):
                    hour+=1
                    mine=0
                if(hour == 24):                   
                    hour=0
                    t=1
                    break
with open('week3_2.csv', "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerows(dataset)
    f.close()
```
